Unemployment is measured by the unemployment rate which is the number of people
who are unemployed as a percentage of the total labour force. We have seen a sharp
increase in the unemployment rate during Covid-19, so analyzing the unemployment rate
can be a good data science project.

```python
# importing all the necessary librarier for dataset 

import pandas as pd
import numpy as np
import matplotlib.pyplot as pl
import seaborn as sn
```

```python
# reading the dataset

data=pd.read_csv("Unemployment in India.csv")
data
```

```python
#Searching for null values form the dataset

data.isnull().sum()
```

```python
# displaying the columns from the dataset

data.columns
```

```python
# Renaming the columns of the dataset for efficiently working with the dataset

data.columns=['Region','Date','Frequency','Estimated Unemployment Rate',
       'Estimated Employed', 'Estimated Labour Participation Rate',
       'Area']
```

```python
# getting teh statistical information about the dataset

data.describe()
```

```python
# Removing the null values from the dataset by droping the nan rows 

data.dropna(inplace=True)
```

```python
# checking the dataset after removing the null values 

data.isnull().sum()
```

```python
# checking for no of rural and urban areas in india

data.groupby("Area").size()
```

```python
# getting the states from the dataset

data.Region.unique()
```

```python
# scatter plot for displaying the unemployment rate 

x=data["Estimated Unemployment Rate"]
y=data["Estimated Labour Participation Rate"]
pl.scatter(x,y)
pl.xlabel("Estimated Unemployment Rate")
pl.ylabel("Estimated Labour Participation Rate")
pl.title("Unemployment Rate")
pl.show
```

```python
# displaying the unemployment rate region wise using the histogram

x=data["Region"]
y=data["Estimated Unemployment Rate"]
pl.ylabel("Region")
pl.xlabel("Estimated Unemployment Rate")
pl.barh(x,y,data=data)
fig=pl.figure(figsize=(10,20))
pl.show()
```

```python
# pie plot for visualizing the distrubution of urban and rural areas

data["Area"]=data["Area"].replace({"Rural":0,"Urban":1})
plabel=["Rural","Urban"]
ex=[0,0]
x=data.groupby("Area").size()
pl.pie(x,labels=plabel,explode=ex)
pl.show()
```
